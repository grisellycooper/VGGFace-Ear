## Data Augmentation

```python
import os
import imageio
import numpy as np
import imgaug as ia
import imgaug.augmenters as iaa
import cv2

from glob import glob
```

```python
ROOT_DIR = os.path.abspath("../../")
DS_DIR = os.path.join(ROOT_DIR, "Datasets/uerc_temp/")
FINAL_DIR = os.path.join(ROOT_DIR, "Datasets/uerc_temp_da/")
if not os.path.exists(FINAL_DIR):
    os.mkdir(FINAL_DIR)

## TO CHECK!! the image file extension .png .jpg .jpeg
    

##### AMI
# Cada clase presenta 7 imagenes, con su respectivo espejo, serian 14. 
# Aplicando un factor de 25, se obtiene 350 imagenes por sujeto

##### WPUT
# Cada clase presenta un promedio de 7 imagenes, con su respectivo espejo, serian 14. 
# Aplicando un factor de 25, se obtiene 350 imagenes por sujeto

aug_factor = 0
final_size_per_class = 60
```

```python
# Flip images and Augmentation
ia.seed(1)

fliponly = iaa.Sequential(
    [
        iaa.Fliplr(1)
    ]
)
### Sometimes(0.5, ...) applies the given augmenter in 50% of all cases,
sometimes = lambda aug: iaa.Sometimes(0.5, aug)
seldom = lambda aug: iaa.Sometimes(0.2, aug)

### Define our sequence of augmentation steps that will be applied to every image.
aug_set = iaa.Sequential(
    [
        # Apply the following augmenters to most images.
        
        # crop some of the images by 0-10% of their height/width
        #iaa.Crop(percent=(0, 0.1)),
        # Blur each image with varying strength using
        # gaussian blur (sigma between 0 and 2.0),
        seldom(iaa.GaussianBlur((0, 1.5))),
        
        # Apply affine transformations to some of the images
        # - scale to 90-110% of image height/width (each axis independently)
        # - rotate by -35 to +35 relative to height/width (per axis)        
        sometimes(iaa.Affine(
            scale={"x": (0.9, 1.1), "y": (0.9, 1.1)},
            translate_percent={"x": (-0.1, 0.1), "y": (-0.1, 0.1)},
            rotate=(-15, 15),            
        )),
        
        # Add gaussian noise to some images. scale=(0, 0.1)
        sometimes(iaa.AdditiveGaussianNoise(scale=(0, 0.2))),
        # Change brightness of images (10-110% of original value).
        sometimes(iaa.Multiply((0.7, 1.3), per_channel=0.3)),
        # Improve or worsen the contrast of images.
        seldom(iaa.ContrastNormalization((0.7, 1.3), per_channel=0.3))
        
    ],
    # do all of the above augmentations in random order
    random_order=True
)

for (dirpath, u_dirnames, filenames) in os.walk(DS_DIR):
    dirnames = sorted(u_dirnames)
    
    for i in range(0,len(dirnames)):
        
        FILE_DIR = os.path.join(DS_DIR, str(dirnames[i]))
        ufiles = glob(FILE_DIR+'/*.png') 
        files = sorted(ufiles)
        
        for j in range (0,len(files)): 
            #print(files[j])
            image = imageio.imread(files[j])
            images=[image]
            images_aug = fliponly(images=images)
            #ia.imshow(np.hstack(images_aug))
            for a in range(len(images_aug)):                
                name = str(files[j][:-4]) + "_flip.png"
                imageio.imwrite(os.path.join(FILE_DIR, name), images_aug[a])
        
        NEW_FILE_DIR = os.path.join(FINAL_DIR, str(dirnames[i]))
        if not os.path.exists(NEW_FILE_DIR):
            os.mkdir(NEW_FILE_DIR)
        
        ufiles = glob(FILE_DIR+'/*.png') 
        files = sorted(ufiles)
        
        if final_size_per_class != 0:
            aug_factor = np.round(final_size_per_class/len(files))
                     
        for j in range (0,len(files)): 
            image = imageio.imread(files[j])
            images=[image] * (aug_factor.astype(int))
            images_aug = aug_set(images=images)
            #ia.imshow(np.hstack(images_aug))
            for a in range(len(images_aug)):                
                name = str(j) + "_" + str(a) + ".png"       
                images_aug[a] = images_aug[a].astype("uint8")
                cv2.imwrite(os.path.join(NEW_FILE_DIR, name), cv2.cvtColor(images_aug[a], cv2.COLOR_RGB2BGR))
        
        print(str(dirnames[i])+ " done!")
        
    break
            
```
